```python
import pandas as pd
```

```python
# reading csv file
df_population_raw = pd.read_csv('population_total.csv')
```

```python
df_population_raw
```

```python
df_population_raw.dropna(inplace=True)
```

```python
df_population = df_population_raw.pivot(
                index='year',
                columns='country',
                values='population'
)
```

```python
df_population = df_population[['United States','India','Brazil','China','Spain']]
df_population
```

### Lineplot

```python
df_population_line = df_population.head()
df_population_line.plot(kind='line')
```

```python
# with label parameters
df_population_line.plot(kind='line', xlabel='Year', ylabel='Population', title='Population Per country')
```

```python
# with figsize parameter (to adjust the size)
df_population.plot(kind='line',figsize=(10,6), xlabel='Year', ylabel='Population',title='Population per country')
```

### Barplot

```python
# Selecting only one year.
df_population_1955  = df_population[df_population.index.isin([1955])]
df_population_1955
```

```python
df_population_1955
```

```python
# Bar Plot (without transposing dataframe)
df_population_1955.plot(kind='bar')
```

```python
# Bar plot with transposing (this is more defined as compared to the non transposed df)
df_population_1955_T = df_population_1955.T
df_population_1955_T
```

```python
df_population_1955_T.plot(kind='bar')
```

#### bar plot's parameters

```python
df_population_1955_T.plot(kind='bar', color='blue',xlabel='Country', ylabel='Population',title='Population(1955)')
```

### Bar Graph by  'n' variable

```python
# df containing only 5 years
df_population_5 = df_population[df_population.index.isin([2000,2005,2010,2015,2020])]
```

```python
df_population_5.plot(kind='bar', xlabel='Year', ylabel='Popluation', title='Population (2000-2020)')
```

### Pie Chart

```python
#creating df with year col
df_pie = df_population[df_population.index.isin([1955])]
df_pie_T = df_pie.T
```

```python
df_pie_T
```
