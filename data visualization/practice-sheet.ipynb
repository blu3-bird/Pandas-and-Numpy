### Exercises to practice data visualization with pandas

```python
import pandas as pd
```

```python
df_population = pd.read_csv('population_total.csv')
df_population.dropna(inplace=True)
df_population.head()
```

Exercise 1 — Create a Pivot Table

Create a pivot table where:

year is the index

country becomes columns

values = population

```python
df_1 = df_population.pivot(index='year',columns='country', values='population')
# selecting only few country
df_1[['United States', 'India','Brazil','Spain', 'China']]
```

Exercise 2 — Line Plot for One Country

Plot a line graph showing the population trend of China over the years.

```python
#2
df_2 = df_1['China']
df_2.dropna()

df_2.plot(kind='line', ylabel='Population', xlabel='China', title='Population over the years in china')
```

Exercise 3 — Compare Two Countries Using a Line Plot

Pick any two countries and plot both population trends together in one line plot.

Add:

title

axis labels

legend

```python
#3 
df_3 = df_1[['China','India']] 
df_3
```

```python
# line plot of two countries
df_3.plot(kind='line',title='Comparing two countries',xlabel='Country', ylabel='Population', legend=True)
```

Exercise 4 — Single Bar Plot

Create a bar plot showing population of India for the last 5 years.

```python
#4
# creating dataframe for last 5 years
df_4 = df_1.loc[2016:2025,['India']]
df_4
```

```python
#single bar plot
df_4.plot(kind='bar',ylabel='Population',xlabel='India',title='Population in India (2016-2020)')
```

Exercise 5 — Grouped Bar Chart

Create a grouped bar chart comparing India vs Pakistan population for the same years.

```python
#5
#creating pivot table 
df_5 = df_1.loc[2016:2020,['India','Pakistan']]
df_5
```

```python
# grouped bar chart
df_5.plot(kind='bar',xlabel='Countries', ylabel='Population',title='Grouped Bar chart')
```
