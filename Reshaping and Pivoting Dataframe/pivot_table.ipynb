## .pivot_table() Method

create a spreadsheet-style pivot table (Supports data aggregation) Similar to MS Excel pivot tables

```python
import pandas as pd
print(pd.__version__)
```

```python
df_sales = pd.read_excel('supermarket_sales.xlsx')
```

```python
df_sales.head()
```

```python
df_sales.select_dtypes(include='number').pivot_table(index=df_sales['Gender'],
                     aggfunc='sum',).round(1)
```

```python
# make a pivot table add an aggfunc and select some cols
df_sales.pivot_table(index='Gender',
                values=['Total','Quantity'],
                aggfunc='sum')
```

```python
# make a pivot table, that displays how much males and females spent in each category
df_sales.pivot_table(
                index='Gender',
                columns='Product line',
                values='Total',
                aggfunc='sum'
)
```
