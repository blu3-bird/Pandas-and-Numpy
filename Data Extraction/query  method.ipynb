```python
import pandas as pd
```

```python
df = pd.read_csv('players_20.csv')
df.set_index('short_name', inplace=True)
df = df.iloc[:,2:9]
df.head()
```

### query method()

```python
# syntax = df.query("condition")
df.query("age>32").head()
```

```python
# players older than 32 and belong to the england
df.query("age>32 and nationality=='England'").head()
```

### 'Not Operator'

```python
# player older than 32 using not operator
df.query("not age<=32").head()
```

```python
# Convert height to meter and select those wwith height above 1.8
df.query("height_cm/100>1.8").head()
```

#### .dtypes function

#### .astype() function
function to change the data type

```python
df['dob'] = df['dob'].astype('datetime64[us]') # dob is datetime type
# extract year from dob
df.query("dob.dt.year > 1990").head()
```
