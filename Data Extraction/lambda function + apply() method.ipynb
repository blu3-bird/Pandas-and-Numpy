```python
import pandas as pd
```

```python
df = pd.read_csv('players_20.csv')
df.set_index('short_name', inplace=True)
df = df.iloc[:,2:9]
df.head()
```

### Lambda function + apply() method

```python
#basic function
def sum_values(a,b):
    return a+b
    
```

```python
sum_values(6,8)
```

```python
# Lambda function
sum_values_lambda= lambda a,b:a+b
sum_values_lambda(7,8)
```

### Lambda function + apply() function

```python
# use lambda function to convert height_cm to meters
df['height_cm'].apply(lambda x: x/100)
```

```python
#Use lambda function to convert 'long_name' to upper_case
df['long_name'].apply(lambda x:x.upper())
#alternative way to get the same output but without lambda method
df['long_name'].str.upper()
```

```python
#use Lambda function to get year from 'dob' series
df.dtypes
df['dob'] = df['dob'].astype('datetime64[us]')
df.dtypes
df['dob'].apply(lambda x: x.year)
```

```python
# apply lambda function to calculate the bmi
df.apply(lambda x:x['weight_kg']/(x['height_cm']/100)**2, axis=1)
```
