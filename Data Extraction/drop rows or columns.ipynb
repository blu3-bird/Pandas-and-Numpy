```python
import pandas as pd
```

```python
df = pd.read_csv('players_20.csv')
df.set_index('short_name', inplace=True)
df = df.iloc[:, 2: 9]
df.head()
```

### Drop method


#### -  Drop Rows 


```python
# two different methods 1. Axis parameter 2. index
df.drop('L. Messi', axis=0).head()
```

```python
# index 
df.drop(index='L. Messi').head()
```

```python
# Multiple rows drop
df.drop(['L. Messi', 'J. Oblak'], axis=0).head()
```

```python
# index
df.drop(index=['L. Messi', 'J. Oblak']).head()
```

#### Drop Columns


```python
#axis
df.drop('long_name', axis=1).head()
```

```python
df.drop(columns='long_name').head()
```

```python
# Drop multiple cols
df.drop(['long_name', 'age', 'dob'],axis=1).head()
```

```python
df.drop(columns=['long_name','age','club']).head()
```
