```python
import pandas as pd
```

```python
df = pd.read_csv('players_20.csv')
df.set_index('short_name', inplace=True)
df = df.loc[:, 'long_name':'club']
df.head()
```

### Set value to one cell

syntax= first select one column/row/cell ,then set its vlaues

```python
# update messi height to 176cm
df.loc['L. Messi', 'height_cm']= 176
df.head(1)
```

### Set value to an entire row


```python
# set height of all player to 180cm
df.loc[:,'height_cm']= 180
df.head()
```

### Set value to entire row

```python
# Set null to all columns of last ranked player in df
import numpy as np
np.nan

df.iloc[-1] = np.nan # instead of using -1 we can simply set the index
df.tail()
```

### Set value to Multiple cells

```python
# set height of two players to 175cm
df.iloc[[0,1], 3] = 175 # here we can either select multiple cols as well
df.head()
```

### set value for rows matching a conditions

```python
# set the value for these cols = age, height, weight and club for those player whose height is more than 180
cols = ['age', 'height_cm', 'weight_kg', 'club']

df.loc[df['height_cm']> 180, cols] = 180
df.head(10)
```
