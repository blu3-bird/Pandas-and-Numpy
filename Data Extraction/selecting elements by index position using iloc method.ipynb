```python
import pandas as pd
```

```python
df = pd.read_csv('players_20.csv')
df = df.set_index('short_name')
df = df.loc[:, 'age':'club']
df.head()
```

### Selecting single value

###  .iloc() method

syntax = df.iloc[row_position,column_position]

```python
# select the height of the Messi
df.iloc[0,2]
```

```python
# get all the rows inside the height_cm col
df.iloc[:, 2]
```

```python
# get all the cols that correspond to index messi
df.iloc[0,:]
```

### Selecting with list of values

```python
# get all the data about L.messi and Cristiano
df.iloc[[0,1], :]
```

```python
# get the height of L messi and Ronaldo
df.iloc[[0,1],2]
```

```python
# get height and weight of Messi
df.iloc[0,[2,3]]
```

```python
# get height and weight  of messi and ronaldo
df.iloc[[0,1],[2,3]]
```

### Selecting a range of data with a slice

start:end here in iloc methods start is included but end is excluded

```python
# slice col labels age/club ,  players = messi and ronaldo
df.iloc[[0,1],0:7] # either we can define players var
```

```python
# slice index labels (top1 to top10) cols: age to club
df.iloc[0:10,0:7] # same can be done by defining the var
```

### Selecting with conditions

Note: In iloc method, it cannot accept a boolean series, but only a boolean list. so we have to use list() function to convert series to list

```python
# select players with height above 180cm and all cols
df.iloc[list(df['height_cm']>180),:].head() 
```

```python
# select the players who weight is above 80kg and from england
condition1 = df['weight_kg'] > 80
condition2 = df['nationality'] == 'England' # this boolean series
# to convert series to list
list((df['weight_kg']> 80)  & (df['nationality']=='England'))
df.iloc[list((df['weight_kg']> 80) & (df['nationality']=='England')), :]
```
