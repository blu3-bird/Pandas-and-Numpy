### Exercise: Multiple cells, conditions and Lists


```python
# 11) Set the nationality of players whose short_name starts with ‚ÄúM‚Äù to "Unknown".
df_m = df.index.str.startswith('M')
df.loc[df_m,'nationality'] = 'Unknown'
df['nationality'].value_counts()
```

```python
#12 
# For Players from brazil: height_cm +=2; weight_kg+=1
df_brazil  = df['nationality'] == 'Brazil'
df.loc[df_brazil,'height_cm'] = df.loc[df_brazil, 'height_cm'] + 2
df.loc[df_brazil,'weight_kg'] = df.loc[df_brazil,'weight_kg'] + 1
df.head()
# Neymar Jr actually height is 175 in df.
```

13) For players younger than 20, set:
