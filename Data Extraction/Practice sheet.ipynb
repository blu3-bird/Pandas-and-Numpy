Practice Sheet (lambda Function)

```python
import pandas as pd
```

```python
df = pd.read_csv('players_20.csv')
df.set_index('short_name',inplace=True)
df=df.iloc[:,2:9]
```

```python
df.head()
```

 Exercise #1:
    -Task: Create a new column 'height_m' by dividing height_cm by 100


```python
col_1 = df['height_cm'].apply(lambda x: x/100)
df.insert(1,'height_m',col_1)
```

```python
# added new col at index 1
df.head()
```

```python
df['height_m'] = df['height_cm'].apply(lambda x: x/100)
df.head()
```

### Exercise 2: Conditional Logic

Task: Create 'age_group' column:
 
 - 'Young' if age < 23

 - 'Prime' if age between 23-30

 - 'Veteran' if age > 30


```python
df['age_group'] = df['age'].apply(lambda x:"Young" if x<23 else("Prime" if x<=30 else "Veteran"))
```

```python
df.head()
```

### Exercise 5: Boolean Mapping
    - Task: Create 'is_tall' column (True if height_cm > 180, False otherwise)


```python
df['is_tall'] = df['height_cm'].apply(lambda x: True if x> 180 else False)
```

```python
df.head()
```

## Intermediate Exercise

### Exercise 1.0: Multiple Column Calculations

 Task: Create 'bmi_category' based on BMI:

 - 'Underweight': BMI < 18.5

 - 'Normal': BMI 18.5-24.9

 - 'Overweight': BMI 25-29.9

 - 'Obese': BMI >= 30


```python
df['bmi'] = df.apply(lambda x:x['weight_kg']/ (x['height_cm']/100) ** 2,axis=1)
```

```python
df.head()
```

```python
df['category_bmi'] = df['bmi'].apply(lambda x:'Underweight' if x < 18.5 else('Normal' if x<=24.9 else('Overweight' if x<=29.9 else 'Obese')))
```

```python
df
```

### Exercise 2: String Manipulation

    - Task: Create 'name_word_count' that counts words in long_name


```python
df['name_word_count'] = df['long_name'].apply(lambda x: len(x.split()))
```

```python
df.head()
```

### Exercise 3: Complex Conditional

 Task: Create 'recommendation' column:

    - 'Top Priority' if age < 25 and height_cm > 180

    - 'Consider' if age < 28 and weight_kg < 80

    - 'Pass' otherwise


```python
df['recommendation'] = df.apply(lambda x: 'Top Priority' if x['age'] < 25 and x['height_cm']>180 else ('Consider' if x['age']<28 and x['weight_kg'] <80 else 'Pass'), axis=1)
```

```python
df
```

### Exercise 4: Working with dates

Calculate years until retirement (assume retirement at 35)

Create 'years-to-retirement' col

```python
df['year_to_retirement'] = df['age'].apply(lambda x: max(0,35-x))
```

```python
df
```

### Exercise 5: String Search

 Task: Create 'is_south_american' (True for Brazil, Argentina, Uruguay, Colombia, Chile)


```python
american = ['Brazil', 'Argentina', 'Uruguay', 'Colombia', 'Chile']

df['is_south_american'] = df['nationality'].apply(lambda x:'True' if x in american else 'False')
```
