```python
import pandas as pd
```

```python
df = pd.read_csv('players_20.csv')
df.set_index('short_name', inplace=True)
df = df[['long_name','age','dob','height_cm','weight_kg','nationality','club']]
df.head()
```

### .loc() method

#### Selecting elements using loc method

```python
# select all data of L.Messi
df.loc['L. Messi']
```

```python
# get the height of the L. Messi	
df.loc['L. Messi', 'height_cm']
```

```python
# to get all col elements of one col
df.loc[:,'weight_kg'] 
```

```python
# to select all the row elements in one row
df.loc['E. Hazard', :]
```

### Select with a list of values

```python
# To select two indexes
df.loc[['E. Hazard', 'L. Messi']]
```

```python
# To select the height of two indexes
df.loc[['E. Hazard', 'L. Messi'], 'height_cm']
```

```python
# to get height and weight of two players.
df.loc [['L. Messi', 'E. Hazard'], ['weight_kg', 'height_cm']]
```

### Select the data using slice

slice : is use to select rows/cols, start:stop both start and stop are included. 

```python
# To select the cols using slice
player = ['L. Messi', 'E. Hazard']
df.loc[player, 'age':'club']
```

```python
# To select the data from rows using slice
# get top 5 players

top = df.index[:10]
df.loc[top, 'age':'club']
```

### Selecting data with conditions

```python
# select the data of the players who's height is more that 180cm
df.loc[df['height_cm']> 180, 'age':'club']
```

```python
# select the players who's weight is more than 75 kg and from Germany
df.loc[(df['weight_kg']>75) & (df['nationality']== 'Germany'), 'age':'club']
```
