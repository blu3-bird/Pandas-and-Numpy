### Exercise: Conditional filtering

9Ô∏è‚É£Select all players taller than 190 cm.

üîü Select all Brazilian players who weigh less than 70 kg.

1Ô∏è‚É£1Ô∏è‚É£ Find all Argentinian players who are older than 30.

1Ô∏è‚É£2Ô∏è‚É£ Select all players with height between 175 and 180 cm.

```python
# Exercise9:
df.loc[df['height_cm']>190]
```

```python
# Exercise 10:
df.loc[(df['nationality']== 'Brazil') & (df['weight_kg']<70)]
```

```python
#Exercise 11:
df.loc[(df['nationality']=='Argentina') & (df['age']> 30)]
```

```python
# Exercise 12:
df.loc[(df['height_cm']>=175) & (df['height_cm']<=180)]
```

### Exercise: Combined Conditions


1Ô∏è‚É£3Ô∏è‚É£ Select all English players who are under 25 years old and play for Manchester United.

1Ô∏è‚É£4Ô∏è‚É£ Select all German or Spanish players taller than 185 cm.

1Ô∏è‚É£5Ô∏è‚É£ Show players whose weight is greater than 90 kg but height is under 185 cm.

```python
# Exercise 13:
df.loc[(df['nationality']=='England') & (df['age']< 25) & (df['club']== 'Manchester City')]
```

```python
# Exercise 14:
df.loc[(df['nationality']=='Germany')| (df['nationality']=='Spain') & (df['height_cm']>185)]
```

```python
# Exercise 15:
df.loc[(df['weight_kg']>90) & (df['height_cm']<185)]
```
