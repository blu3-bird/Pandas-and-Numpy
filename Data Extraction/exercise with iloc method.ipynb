### Exercise: Boolean conditions with iloc[]
21. Select players taller than 185 cm

22. Select players younger than 20 AND weigh less than 70 kg

23. Select players from Argentina OR Brazil


```python
#21
df.iloc[list(df['height_cm']>185),:]
```

```python
#22
df.iloc[list((df['age']<20) & (df['weight_kg']<70)),:]
```

```python
#23
df.iloc[list((df['nationality'] == 'Argentina') | (df['nationality']=='Brazil')),:]  # correct but same result can be obtained by using isin()
df.iloc[list(df['nationality'].isin(['Argentina', 'Brazil']))]
```

### Exercise: Combine slicing + boolean masking




```python
# From the first 500 rows, select rows where:

#   height > 175

#   AND weight < 75

#   Only return columns age, dob, height, weight
```

```python
df_500 = df.head(500)
df_500.iloc[ list((df_500['height_cm']> 175) & (df_500['weight_kg']<75))]
```
