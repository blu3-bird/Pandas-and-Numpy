### Copy method

1.1 Deep parameter = True


```python
# df_copy is copy of df. with default value of deep = True, any change to this file will not be reflected to copied file
df_copy = df.copy()
```

1.2 Deep parameter = False

```python
# with deep = false, any change to original file will reflect in the copy df
df_copy_new = df.copy(deep=False) 
```

```python
df.iloc[1,2] = 34

# copied df
df_copy_new.head()
```

1.3 Simple assignment

```python
#  simple way to make copy but here any change made to the parent file will be reflect in the copied file 
df_copy_3 = df
```

```python
# change made to any of the file will be reflect in other one, in simple words, both files are independent on each other.
df_copy_3.iloc[1,3]=50
df.head()
```
