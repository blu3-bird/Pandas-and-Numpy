```python
import numpy as np
import pandas as pd
import random as rd
```

```python
df_exams = pd.read_csv('StudentsPerformance.csv')
df_exams.head()
```

### 1.1 Create index

```python
# Create an array using np
new_index = np.arange(0,1000)
len(new_index)
```

```python
# shuffling the array
rd.shuffle(new_index)
new_index
```

```python
# Adding new col in df
df_exams['new_index'] = new_index
```

```python
# displaying the df
df_exams.head()
```

### 1.2 Set index() method

```python
# Set new add col as index of DF
df_exams.set_index('new_index')
```

### 1.3 Sort Index() method

```python
# sort dataframe by new index
df_exams.sort_index()
```

### 2. Rename Column

```python
# Rename col and overwrite values of the dataframe
df_exams = df_exams.rename(columns={'gender': 'Gender'})
df_exams.head()
```

```python
# Renaming multiple columns 
df_exams.rename(columns={'math score': 'MS', 'writing score': 'WS', 'rs': 'RS'}, inplace=True)
```

```python
df_exams.head()
```

### 3 Rename Index

```python

df_exams = df_exams.sort_values('new_index')
df_exams.head()
```

```python
# rename the index
df_exams.rename(index={0: 'A', 1: 'a', 2: 'B'}, inplace=True)
df_exams.head()
```
