### 1.2 Reading multiple csv from URL 

link structure
https://www.football-data.co.uk/mmz4281/2526/E0.csv

root(https://www.football-data.co.uk/mmz4281/) + 'league year'+'/'+league series(E0)+ '.csv'

```python
# creating root variable
root = 'https://www.football-data.co.uk/mmz4281/'
```

```python
# create a for loop to loop through leauge series
leagues = ['E0', 'E1', 'E2', 'E3', 'EC']
frames = [] # empty dict to store each df

for league in leagues:
    df = pd.read_csv(root+ '2526'  + '/' + league + '.csv')
    frames.append(df)
```

```python
# length of frames
len(frames)
```

```python
# showing the elements of frames
frames[0].head()
```

### 1.3 csv file from multiple seasons

```python
# looping through the leagues and seasons
leagues = ['E0', 'E1', 'E2', 'E3', 'EC']
frames = []

for league in leagues:
    for i in range(19,24):
        season = str(i)+ str(i+1)
        df = pd.read_csv(root + season + '/' + league + '.csv', encoding='unicode_escape')
        df.insert(1,'Season', season)
        frames.append(df)
```

```python
len(frames)
```

```python
frames[24].head()
```

### Organizing Data in dict

```python
dict_countries = {
              'Spanish La Liga':'SP1', 'Spanish Segunda Division':'SP2',
              'German Bundesliga':'D1', 
              'English Premier League':'E0', 'English League 1':'E2', 'English League 2':'E3',
             }
```

```python
dict_countries['English League 2']
```

```python
dict_all_data = {}

for a in dict_countries:
    frames = []
    for season in range(20,24):
        year = str(season) + str(season+1)
        dt = pd.read_csv(root+ year + '/' + dict_countries[a] + '.csv')
        dt.insert(1, 'Season', year)
        frames.append(dt)
    f_df = pd.concat(frames)
    dict_all_data[a] = f_df
```
