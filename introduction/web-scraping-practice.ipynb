### Scraping data of different seasons and leagues


Multiple seasons csv file links:

https://www.football-data.co.uk/mmz4281/2526/E0.csv

https://www.football-data.co.uk/mmz4281/2425/E0.csv

https://www.football-data.co.uk/mmz4281/2223/E2.csv

```python
# creating Year variable to store different seasons
bird = []
for season in range(20,23):
    year = f'{season}{season+1}'
    bird.append(year)
    print(bird)
```

```python
Arson = [] 
leagues = ['E0', 'E1', 'E2']
for league in leagues:
    for i in bird:
        dataSet = pd.read_csv(Root + i + '/' + league + '.csv', encoding='unicode_escape')
        dataSet.insert(1,'Season', i)
        Arson.append(dataSet)
        
    
len(Arson)
```

### Organizing data in dict

```python
# create a dictionary with original name of league as key
dict_countries = {
              'Spanish La Liga':'SP1', 'Spanish Segunda Division':'SP2',
              'German Bundesliga':'D1', 
              'English Premier League':'E0', 'English League 1':'E2', 'English League 2':'E3',
             }
```

```python
dict_countries['English Premier League']
```

```python
all_data_dict = {}
for league in dict_countries:
    frames = []
    df = pd.read_csv(Root+ i + '/' + dict_countries[league]+ '.csv')
    #df.insert[1,'Season', year]
    frames.append(df)
    df_frames = pd.concat(frames)
    all_data_dict[league] = df_frames
```
