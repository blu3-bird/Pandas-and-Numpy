```python
import pandas as pd
```

```python
# Reading the csv file
df_exams = pd.read_csv('StudentsPerformance.csv')
df_exams.head()
```

## 1 Math Operations

### 1.1 Operatios in columns

```python
# Select a column and calculate total sum
df_exams['math score'].sum()
```

###### To perform math operations on col, we have to select one first


```python
# Count, mean, std, max and min function
df_exams['math score'].count()
df_exams['math score'].mean()
df_exams['math score'].std()
df_exams['math score'].max()
df_exams['math score'].min()
```

```python
# Calculation with .describe()
df_exams.describe()
```

### 1.2 Operations in Rows

```python
# Calculating the sum in a row
df_exams['math score'] + df_exams['reading score'] + df_exams['writing score']
```

```python
# Calculate the average and assign it to new col
df_exams['Average'] = (df_exams['math score'] + df_exams['reading score']+ df_exams['writing score'])/3
```

```python
df_exams.head().round(2)
```

## 2 Value Counts

```python
# counting gender elements

# len function
len(df_exams['gender'])

# .count() method
df_exams['gender'].count()
```

#### 2.1 counting elements by category

```python
df_exams['lunch'].value_counts()
```

```python
# return the relative frequency (divide all values by the sum of values)
df_exams['gender'].value_counts(normalize = True)
```

## 1. Sort a DataFrame

```python
# Sort by one Column
df_exams.sort_values('lunch')
```

```python
# Sort descending by multiple columns and update dataframe
df_exams.sort_values(['math score', 'reading score', 'writing score'], ascending=False, inplace=True)
```
