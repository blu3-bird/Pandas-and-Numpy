```python
import pandas as pd
```

```python
df_laptops = pd.read_csv('laptop_price.csv', encoding='latin1')
```

```python
df_laptops.head()
```

###### pandas syntax 
and -> &
or -> |

### Filtering df based on 2 or more conditions

#### 'And' condition

```python
# conditions: filter the laptops that are macbook and costs more that 2000 euros
(df_laptops['Company'] == 'Apple') & (df_laptops['Price_euros'] <2000)
df_laptops[(df_laptops['Company'] =='Apple') & (df_laptops['Price_euros'] >2000)]
```

### 'Or' Condition

```python
# condition: either 'Dell' or 'Apple' laptops
(df_laptops['Company'] == 'Dell') | (df_laptops['Company'] == 'Apple')
df_laptops[(df_laptops['Company']=='Dell') | (df_laptops['Company']=='Apple')]
```

### Multiple conditons

```python
# Conditon: Find laptops from apple or dell that costs more that 2000 euros
df_laptops[( (df_laptops['Company'] == 'Apple') | (df_laptops['Company'] == 'Dell') ) & (df_laptops['Price_euros'] > 2000)].head()
```
