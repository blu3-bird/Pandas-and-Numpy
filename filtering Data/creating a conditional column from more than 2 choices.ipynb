### Creating a column from more than 2 values

```python
import pandas as pd
import numpy as np
```

```python
df_laptops = pd.read_csv('laptop_price.csv' , encoding='latin1')
```

```python
df_laptops.head()
```

```python
# set up two lists to store conditions and values

conditions = [
    (df_laptops['Price_euros'] >3000),
    (df_laptops['Price_euros'] >2000) & (df_laptops['Price_euros']<= 3000),
    (df_laptops['Price_euros']> 800) & (df_laptops['Price_euros']<=2000),
    (df_laptops['Price_euros']<=800),
]

choice = ["Too Expensive", "Expensive", "Affordable" , "Cheap"
]
```

```python
# set it to a new col
df_laptops['Price_Tier'] = np.select(conditions, choice, default='Some String')
```

```python
df_laptops.head()
```

```python
df_laptops.value_counts('Price_Tier')
```

```python
# create an array based on multiple screen size tiers 
# T

Args = [
    (df_laptops['Inches'] > 16),
    (df_laptops['Inches']< 16) & (df_laptops['Inches'] >= 14),
    (df_laptops['Inches']< 14) & (df_laptops['Inches']>=12),
    (df_laptops['Inches']< 12)
]

keys =[
    'Too Big', 'Big' , 'Mid' , 'Small'
]
```

```python
# set it to new col
df_laptops['Size_Tier'] = np.select(Args, keys, default='Some String')
```

```python
df_laptops.value_counts('Size_Tier')
```
