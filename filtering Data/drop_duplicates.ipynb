## Drop Duplicates() method

### Find cheap and most expensive laptops using .drop_duplicates()  method

```python
# sort the dataframe
df_laptops = df_laptops.sort_values(['Company', 'Price_euros'])
df_laptops.tail()
```

```python
# removing the duplicates in our dataframe ( keep = 'first' ) cheap laptops from each company
df_cheap_laptops = df_laptops.drop_duplicates(['Company'])[['Company', 'Price_euros']]
df_cheap_laptops
```

```python
# removing the duplicates elements from the company col, (keep = 'last') 
df_expensive_laptops = df_laptops.drop_duplicates(['Company'], keep='last')[['Company', 'Price_euros']]
df_expensive_laptops
```

```python
# Ignore_index argument
# by default it is set to false, if we set it to true we will have a new index to #our dataframe

df_ignore_index = df_laptops.drop_duplicates(['Company'], keep = 'last', ignore_index=True)[['Company','Price_euros' ]]
df_ignore_index
```

### Exercise

Task:
Find the biggest and smallest screen size in laptops for each company using sort_values() and duplicated() methods (keep first last and false)

```python
df_laptops.head()
```

```python
# sort the dataframe
df_laptops_sort = df_laptops.sort_values(['Company', 'Inches'])
df_laptops_sort.head()
```

```python
# marking the duplicates elements in company col
df_laptops_size_first = df_laptops_sort.duplicated(['Company'], keep='first')
df_laptops_size_first
```

```python
# filtering the dataframe 
df_laptops_sort[~df_laptops_size_first][['Company', 'Inches']]
# list of smallest laptops from each company
```

### Exercise: filtering biggest laptops

```python
# marking the duplicate from the sort df
df_laptops_size_last = df_laptops_sort.duplicated(['Company'], keep='last')
df_laptops_size_last
```

```python
# filtering the dataframe
df_laptops_sort[~df_laptops_size_last][['Company', 'Inches']]

# biggest screen size from each company 
```

### Solving the same exercise using the drop_duplicates() method

```python
# sorting the df
df_laptops_sort = df_laptops.sort_values(['Company', 'Inches'])
```

```python
# smallest screen size:
df_laptops_size_screen_small = df_laptops_sort.drop_duplicates(['Company'], keep='first', ignore_index=True)[['Company', 'Inches']]
df_laptops_size_screen_small
```

```python
# Biggest screen size:
df_laptops_screen_size_big = df_laptops_sort.drop_duplicates(['Company'], keep='last', ignore_index=True)[['Company', 'Inches']]
df_laptops_screen_size_big
```
