```python
import pandas as pd
```

```python
df_laptops = pd.read_csv('laptop_price.csv', encoding='latin1')
```

```python
df_laptops.head()
```

```python
# Sorting values
df_laptops = df_laptops.sort_values(['Company', 'Price_euros'])
```

```python
df_laptops.head()
```

```python
df_laptops.value_counts('Company')
df_laptops['Price_euros'].min()
```

```python
# Marking duplicated elements in our company col
df_first = df_laptops.duplicated('Company', keep='first')
```

```python
# Filtering duplicated elements only
df_laptops[df_first][['Company', 'Price_euros']]

# Filtering non duplicated elements. ( Cheap laptops list)
df_laptops[~df_first][['Company','Price_euros']]

```

### Exercise 2: filter most expensive laptops from each company

```python
# Reading CSV file
df_laptops = pd.read_csv('laptop_price.csv', encoding='latin1')
```

```python
# Sort the dataframe in descending order
df_laptops = df_laptops.sort_values(['Company', 'Price_euros'], ascending=False)
df_laptops.head()
```

```python
# marking duplicates in company 
df_last = df_laptops.duplicated(['Company'])
df_last
```

```python
# filtering Non duplicated elements ( most expensive laptops from each company )
df_laptops[~df_last][['Company', 'Price_euros']]
```
