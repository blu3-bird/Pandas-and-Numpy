```python
import pandas as pd
```

```python
df_laptops = pd.read_csv('laptop_price.csv', encoding='latin1')
```

```python
df_laptops.head()
```

### Isin() method

#####  it is used to filter df based on conditions

```python
# Condition: Either apple or hp laptop
df_laptops['Company'].isin(['Apple', 'HP'])
```

```python
# Filtering Df based on above condition
df_laptops[df_laptops['Company'].isin(['Apple', 'HP'])].info()
```

### Multiple conditions

```python
# condition: Either Ultrabook or Notebook of HP or Apple laptops

# Define variables for conditions

filter1 = df_laptops['TypeName'].isin(['Ultrabook', 'Notebook'])
filter2 = df_laptops['Company'].isin(['Apple', 'HP'])
```

```python
# Create series of our conditions
filter1 & filter2
```

```python
# filter df based on our conditions
df_laptops[filter1 & filter2].info()
```
