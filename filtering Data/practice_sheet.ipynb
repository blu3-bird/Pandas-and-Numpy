```python
import pandas as pd
```

```python
df = pd.read_csv('laptop_price.csv',encoding='latin1')
df.head()
```

```python
# find the cheap and expensive laptop from each company
df_sort = df.sort_values(by=['Company','Price_euros'])
```

```python
# finding duplicates in company col

df_duplicates = df_sort.duplicated('Company')
```

```python
# dataframe (duplicate values)
df[df_duplicates][['Company','Price_euros']] 
```

```python
# non duplicates dataframe 
"""Cheap Laptops from each company"""
df[~df_duplicates][['Company','Price_euros']]
```

### Most Expensive laptops

```python
# sorting the dataframe in des order
df_desort = df.sort_values(by=['Company', 'Price_euros'], ascending=False)
```

```python
# marking the duplicates
df_desort_dup = df_desort.duplicated('Company')
df_desort_dup
```

```python
# dataframe
df[~df_desort_dup][['Company','Price_euros']]
```

### Most Expensive laptops with keep='last'

```python
df_exp = df_sort.duplicated( 'Company' , keep='last')
```

```python
df[~df_exp][['Company','Price_euros']]
```
